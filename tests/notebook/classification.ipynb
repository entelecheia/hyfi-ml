## Prepare an environment

```python
from hyfiml import HyFI

if HyFI.is_colab():
    HyFI.mount_google_drive()

h = HyFI.initialize(
    project_name="hyfi-ml",
    project_root="$HOME/workspace/projects/hyfi-ml",
    logging_level="INFO",
    verbose=True,
)

print("project_dir:", h.project.root_dir)
print("project_workspace_dir:", h.project.workspace_dir)
```

```python
data_dir = h.project.workspace_dir / 'data'
data_dir
```

## Building `cointax_polarity` dataset

```python
raw_data = HyFI.load_dataframe("cointax.csv", data_dir)
raw_data.info()
```

```python
import numpy as np

data = []
for polarity, rows in raw_data.to_dict().items():
    data += [
        dict(text=text, labels="postive" if polarity == "긍정" else "negative")
        for text in rows.values()
        if text is not np.NaN
    ]
data = HyFI.records_to_dataframe(data)
HyFI.save_dataframes(data, "cointax_rawdata.parquet", data_dir)
data
```
